```python
import numpy as np
from NeuralNetwork import NeuralNetwork
from Layer import Layer
import matplotlib.pyplot as plt
```

```python
input = Layer(2,2)
hidden = Layer(2,8)
output = Layer(8, 1)
nn = NeuralNetwork([input, hidden, output])
```

```python
rand_x = np.random.uniform(-8,8,3000)
rand_y = np.random.uniform(-2,2,3000)
target = np.sin(rand_x) > rand_y
target = [[1] if x == True else [0] for x in target.tolist()]
```

```python
plt.scatter(x=rand_x, y=rand_y, s=10)
```

```python
nn.train(list(zip(rand_x,rand_y)), target, 2000)
```

```python
test_x = np.random.uniform(-8,8,3000)
test_y = np.random.uniform(-2,2,3000)
test = list(zip(test_x, test_y))
```

```python
answer = []
for itest in test:
    answer.append(*[1 if x > 0.5 else 0 for x in nn.predict(itest)])
```

```python
sinx=np.linspace(-8, 8, 80)
siny=np.sin(sinx)
fig, ax = plt.subplots()
ax.scatter(x=test_x, y=test_y, c=answer, s=10)
ax.plot(sinx, siny, c='red', linewidth=2.0)
```
